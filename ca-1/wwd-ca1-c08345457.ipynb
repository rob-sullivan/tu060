## Problem definition
The [european protection agency, 2021](https://www.epa.ie/environment-and-you/climate-change/what-impact-will-climate-change-have-for-ireland/) stated that "human activities are estimated to have caused approximately 1.0°C of global warming above pre-industrial levels with a likely range of 0.8°C to 1.2°C". The EPA believes we are on course to exceed the paris climate agreement of remaining between 1.5°C to 2°C above pre-industrial levels. This will impact ireland from a decrease in frequency of summer rain fall and an increase in of heavy winter rain, it will change wind strength and put stress on crops. Rising sea levels will cause coastal erosion, flooding and damage to both infastructure or propery in all major cities in Ireland that are located near the coast.This would affect the Irish economy, human health and wellbeing.

According to the [Sustainability Energy Authority of Ireland (SEAI), 2021)](https://www.seai.ie/data-and-insights/seai-statistics/key-statistics/co2/) transportation accounts for 40% of our kilo tonne of C02 put into the atmosphere.

Reducing the number of petrol and diesel vehicles on Irish roads could reduce this 40% figure. Ways to reduce the number of vehicles on the road are for commuters to change to electric vehicles, cycle more or use public transport.

This report will look at
- Different modes of transport for people in ireland
- Understanding from the data what makes people choose personal transport over public transport. 
- Investigate, from the data, how to increase public transport usage.

```python
#loop through columns, remove white spaces
#this is to help accessing values later

def format_columns(data_frame):
    formated_names = {}
    for col_name in data_frame.columns:
        name = col_name.lower() # could also have used df = df.rename(str.lower, axis='columns')
        name = name.replace(' ', '_')
        formated_names[col_name] = name
    return data_frame.rename(columns=formated_names)

df = format_columns(df)
df.info()
```

```python
#loop through columns if data type is object convert it to category
def categorise_objects(data_frame):
    for col_name in data_frame.columns:
        dt = data_frame.dtypes[col_name]
        if dt == object:
            data_frame[col_name] = data_frame[col_name].astype('category')
    return data_frame

df = categorise_objects(df)
df.info()
```

```python
#Look to see what unique values are in each column
# idea here is to see what can be dropped and what can be kept
def list_unique_cats(data_frame):
    for col_name in df.columns:
        print("\n" + col_name)
        print(df[col_name].unique())
list_unique_cats(df)
```

```python
    #Statistic column only has 1 value which is not useful to us. We will remove this.
    #Census Year is important it is a date range for 7 different census surveys taken over 30 years
    #Sex 
        # has 3 options one is called 'Both Sexes' 
        # 'Both Sexes'was renamed to all via df = df.replace({'Both sexes': 'all'})
        # when viewing dataset it appears to be a summation of other columns.
        # any 'all' in dataset will be removed.
    #Means of Travel 
        # has 'not stated and 'all means of travel'
        # 'all means of travel' was renamed to all via df = df.replace({'Both sexes': 'all'})
        # any 'all' in dataset will be removed.
        # motor car has both a driver and a passenger option
        # some names could be changed for brevity, e.g 
            # All means of travel -> all
            # Work mainly at or from home -> remote work
            # On foot -> walking
            # bicycle -> bike
            # Bus, minibus or coach -> bus
            # Motorcycle or scooter -> motorbike
            # Motor car: Driver -> car
            # Motor car: Passenger -> carpool
            # Other means (incl. lorry or van) - truck
    #At Work School or College
        # this column shows the level of education a person has
        # but it also contains age data
        # we could seperate these two into a new column age and edu_level
        # this seperation would be based on info available from [citizens information](https://www.citizensinformation.ie/en/education/the_irish_education_system/overview_of_the_irish_education_system.html)
        # 'all persons' could be removed or held as a seperate database
        #'all persons' isn't describe clearly. e.g Is the person has no education or is 89 years old?
        #some names could be changed for brevity as well, e.g
            #'All persons', 
            #'Children at school aged between 5 and 12 years', 5-12 | primary-school
            #'Population aged 15 years and over at work', >=15 | junior-cert
            #'Students at school or college aged between 13 and 18 years', 13-18 | leaving-cert
            #'Students at school or college aged 19 years a...', 19-up | university
    #Unit 
        #supposed to be if the unit of measure is km, mins, etc but for this dataset it is just number
        # since one number we will remove
    #Value 
        #is number of people we will keep  
```

```python
#we can now format columns (e.g drop a column or seperate a column)
df = df.drop('statistic', axis=1) # axis 0 will drop label, 1 will drop column
df = df.rename(columns={'censusyear':'year'})#we can drop census from name of column
df = df.replace({'Both sexes': 'all'}) #we rename 'Both sexes' values to 'all'
#rename means of travel
df['means_of_travel'] = df['means_of_travel'].cat.rename_categories({
    'On foot': 'walking', 
    'Bicycle': 'bike', 
    'Bus, minibus or coach': 'bus', 
    'Train, DART or LUAS': 'train', 
    'Motorcycle or scooter': 'motorbike',
    'Motor car: Driver': 'car',
    'Motor car: Passenger': 'carpool',
    'Other means (incl. lorry or van)': 'truck',
    'Work mainly at or from home' : 'remote',
    'All means of travel' : 'all',
    'Not stated' : 'na'
})
#split out at_work_school_or_college
age = []
edu_level = []
for value in df['at_work_school_or_college']:
    if value == "All persons":
        age.append("all")
        edu_level.append("all")
    elif value == "Children at school aged between 5 and 12 years":
        age.append("5 to 12")
        edu_level.append("primary-school")
    elif value == "Population aged 15 years and over at work":
        age.append(">=15")
        edu_level.append("junior-cert")
    elif value == "Students at school or college aged between 13 and 18 years":
        age.append("13 to 18")
        edu_level.append("leaving-cert")
    elif value == "Students at school or college aged 19 years and over":
        age.append(">=19")
        edu_level.append("university")  
df['age'] = age
df['edu_level'] = edu_level

#we now don't need this column anymore
df = df.drop('at_work_school_or_college', axis=1)

#we can drop units
df = df.drop('unit', axis=1)

#now we remove every 'all' row
df = df[df.sex != 'all']
df = df[df.means_of_travel != 'all']
df = df[df.age != 'all']
df = df[df.edu_level != 'all']

#rearange columns so value is at end
df = df[['year', 'sex', 'means_of_travel', 'age', 'edu_level', 'value']]
```
